```python
from langchain.tools import tool
from langchain.agents import create_agent

@tool
def retriever_tool(user_query: str) -> str:
    """Retrieve relevant document chunks based on the user query Context is related to information about the product 
    like details, technical specifications, features, compatibility, pricing, and availability."""
    retriever = vector_store.as_retriever(search_kwargs={"k": 3})
    chunks = retriever.invoke(user_query)
    return ' '.join([chunk.page_content for chunk in chunks])

agent = create_agent(
    llm, 
    tools=[retriever_tool],
    system_prompt=" You are a helpful assistant ")
```

## Evaluation with DeepEval

```python
eval_questions = open('retreiver_evals.txt').readlines()
```

```python
from deepeval.test_case import LLMTestCase


input_query = eval_questions[0]
test_cases = []
for input_query in eval_questions:
    response = agent.invoke({
        'messages': input_query
    })

    retrieval_context = []
    for message in response['messages']:
        print(message)
        if message.type == 'tool' and message.name == 'retriever_tool':
            retrieval_context.append(message.content)

    final_response = response['messages'][-1].content   

    test_case = LLMTestCase(
        input=input_query,
        actual_output=final_response,
        retrieval_context=retrieval_context
    )     
    test_cases.append(test_case)
```

```python
test_cases
```

```python
import os
from deepeval.models import AzureOpenAIModel
from deepeval.metrics import AnswerRelevancyMetric

model = AzureOpenAIModel(
    model_name=os.environ['AZURE_OPENAI_CHAT_MODEL_NAME'],
    deployment_name=os.environ['AZURE_OPENAI_CHAT_DEPLOYMENT_NAME'],
    azure_openai_api_key=os.environ['AZURE_OPENAI_API_KEY'],
    openai_api_version='2024-08-01-preview',
    azure_endpoint=os.environ['AZURE_OPENAI_ENDPOINT'],
    temperature=0
)
```

```python
from deepeval import evaluate

from deepeval.metrics import FaithfulnessMetric

metric = FaithfulnessMetric(
    threshold=0.7,
    model=model,
    include_reason=True
)

metric = evaluate(test_cases=test_cases, metrics=[metric])
```

```python
for test_case in metric.test_results:
    if test_case.success is False:

        print("Input Query:", test_case.input)
        print("Actual Output:", test_case.actual_output)
        print("Retrieval Context:", test_case.retrieval_context)

        print("-" * 50)
```
